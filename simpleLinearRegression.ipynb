```python
#here we are importing pandas library it contains tables,etc.
import pandas as pd
```

```python
#here in dataset variable we are storing the csv file that is dataset file values.
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#using this dataset varibale we will come to know that csv value is stored succesfully in dataset variable.
dataset
```

```python
#the variable independent indicates that it is input. here the yearsExperience is input.
independent = dataset[["YearsExperience"]]
```

```python
# now using this we will confirm it the i/p is crtly came .
independent
```

```python
#here dependent variable is output variable. salary is o/p.
dependent = dataset[["Salary"]]
```

```python
dependent
```

```python
#here we are using the sklearn and spliting the test and train set.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# here the first and second line is procedure and now we the value are substitued in the paramete the bias and weight will be calculated. 
#once it's calculated the model is ready and created.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#here we will come to know about the weight of the model.
weight = regressor.coef_
weight
```

```python
#here we will come to know about the bias of the model.
bias = regressor.intercept_
bias
```

```python
#here in regressor the model will be saved and using predict function and X_test is test data set. using it we wil predict the
# o/p
y_pred=regressor.predict(X_test)
```

```python
#here we are importing the evaluting metrices and subs in r2 it will calculate the score and give.
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#save the model
import pickle
filename ="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
